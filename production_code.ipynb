## Search and Choose Function

```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display

def filter_data(df, cable_type=None, length=None, color=None):
    filtered_data = df.copy()

    if cable_type:
        filtered_data = filtered_data[filtered_data['Type'] == cable_type]
    if length:
        filtered_data = filtered_data[filtered_data['Length'].astype(str) == length]
    if color:
        color = color.lower()  # Convert user input to lowercase
        filtered_data = filtered_data[filtered_data['Color'].astype(str).str.lower() == color]

    return filtered_data

def main():
    df = pd.read_csv('production.csv')

    # Get unique values for cable type, color, and length
    unique_types = [''] + sorted(df['Type'].unique().tolist())
    unique_colors = [''] + sorted(df['Color'].astype(str).unique().tolist())
    unique_lengths = [''] + sorted(df['Length'].astype(str).unique().tolist())

    # Create dropdown widgets
    cable_type_dropdown = widgets.Dropdown(options=unique_types, description="Cable Type:")
    color_dropdown = widgets.Dropdown(options=unique_colors, description="Color:")
    length_dropdown = widgets.Dropdown(options=unique_lengths, description="Length:")

    # Display dropdown widgets
    display(cable_type_dropdown)
    display(color_dropdown)
    display(length_dropdown)

    def on_button_click(b):
        # Get selected values from dropdowns
        cable_type = cable_type_dropdown.value if cable_type_dropdown.value else None
        color = color_dropdown.value if color_dropdown.value else None
        length = length_dropdown.value if length_dropdown.value else None

        # Filter data based on user input
        filtered_df = filter_data(df, cable_type, length, color)

        # Display results
        print("\nFiltered Data:")
        print(filtered_df)

        if filtered_df.empty:
            print("No matching entries.")

    # Create a button to trigger filtering
    filter_button = widgets.Button(description="Filter Data")
    filter_button.on_click(on_button_click)
    display(filter_button)

if __name__ == "__main__":
    main()
```

## Total Time/person Function

```python
import pandas as pd
from tabulate import tabulate

columns = ["Date", "Type", "Color", "Length", "Number", "Production_number", "Steps", 
           "Abnormal", "Mistake_rates", "People", "Total_time", "Time_per_person"]

# Convert "Total_time" and "Time_per_person" to numeric
df["Total_time"] = pd.to_numeric(df["Total_time"], errors="coerce")
df["Time_per_person"] = pd.to_numeric(df["Time_per_person"], errors="coerce")

# Group by Type, Color, and Length
grouped_df = df.groupby(["Type", "Color", "Length"])

# Store the results in a list of tables
tables = []

# Display the results
for name, group in grouped_df:
    # Calculate the total time per person for each group
    total_time_per_person = group["Time_per_person"].sum()
    
    # Get the last row of the Steps column for each group
    last_step = group.iloc[-1]["Steps"]
    
    # Add data to the table
    result_table = {
        "Type": name[0],
        "Color": name[1],
        "Length": name[2],
        "Total_time_per_person": total_time_per_person,
        "Last_step": last_step,
    }
    
    # Append the result table to the list
    tables.append(result_table)

# Convert the list of tables to a DataFrame
result_df = pd.DataFrame(tables)

# Display the final DataFrame
print(tabulate(result_df, headers='keys', tablefmt='pretty', showindex=False))
```

## Test within a week 

```python
import pandas as pd
from tabulate import tabulate

# Define column names
columns = ["Date", "Type", "Color", "Length", "Number", "Production_number", "Steps", 
           "Abnormal", "Mistake_rates", "People", "Total_time", "Time_per_person"]

# Convert "Total_time" and "Time_per_person" to numeric
df["Total_time"] = pd.to_numeric(df["Total_time"], errors="coerce")
df["Time_per_person"] = pd.to_numeric(df["Time_per_person"], errors="coerce")

# Convert "Date" column to datetime format
df["Date"] = pd.to_datetime(df["Date"])

# Group by Type, Color, and Length
grouped_df = df.groupby(["Type", "Color", "Length"])

# Display the results
for name, group in grouped_df:
    print("\nType: {}\nColor: {}\nLength: {}".format(*name))
    
    # Ensure each subgroup is within 7 days
    subgroups = []
    current_subgroup = []
    current_subgroup_duration = pd.Timedelta(0)
    
    for _, row in group.iterrows():
        if not current_subgroup or (row["Date"] - current_subgroup[-1]["Date"]).days <= 7:
            current_subgroup.append(row)
            current_subgroup_duration += pd.to_timedelta(row["Time_per_person"], unit="h")
        else:
            subgroups.append(current_subgroup)
            current_subgroup = [row]
            current_subgroup_duration = pd.to_timedelta(row["Time_per_person"], unit="h")
    
    # Add the last subgroup
    subgroups.append(current_subgroup)
    
    # Display subgroups
    for i, subgroup in enumerate(subgroups):
        print(f"Subgroup {i + 1}:")
        table = tabulate(pd.DataFrame(subgroup)[columns], headers="keys", tablefmt="pretty", showindex=False)
        print(table)
        subgroup_total_time = sum(row["Time_per_person"] for row in subgroup)
        print(f"Subgroup {i + 1} Total Time: {subgroup_total_time} hours")
    
    print("=" * 50)
```
